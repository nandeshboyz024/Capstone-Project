## REMOVE AND FILL UNKNOWN VALUES

```python
# is data as 'unknown'  on 'checking_balance' 
check_unknown = pd.DataFrame(credit_data.checking_balance=='unknown')  # if condition is satisfied check_unknown store True else False

credit_data[check_unknown['checking_balance'] == True]   # from temp take only those rows where checking_balance has True
```

```python
credit_data.mode() 
```

```python
credit_data['checking_balance'].mode() 
# here the unknown value is mode of this column which would be replaced by mode . but in this case it is replace by second mode of this column.
```

```python
credit_data['checking_balance'].value_counts()   # count all value in this column 
# second mode of this column is < 0 DM which has frequency of 274 .
```

```python
credit_data['savings_balance'].mode()
```

```python
# Missing values have a'unknown''
# Replace missing values with NaN
credit_data =credit_data.replace('unknown', np.nan)
credit_data[check_unknown['checking_balance']== True] 
```

```python
credit_data['checking_balance'] = credit_data['checking_balance'].replace(np.nan, '< 0 DM ')
credit_data['savings_balance'] = credit_data['savings_balance'].replace(np.nan, '< 100 DM ')
```
